```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from google.colab import files
from io import BytesIO
from PIL import Image

model = keras.applications.VGG16(
    include_top=True,
    weights="imagenet",
    input_tensor=None,
    input_shape=None,
    pooling=None,
    classes=1000, # output
    classifier_activation="softmax",
)

uploaded = files.upload()
img = Image.open(BytesIO(uploaded['ex224.jpg']))
plt.imshow( img )

# приводим к входному формату VGG-сети
img = np.array(img)
# BGR
x = keras.applications.vgg16.preprocess_input(img)
print(x.shape)
# output format: (batch_size, height, width, channels)
x = np.expand_dims(x, axis=0)

# прогоняем через сеть
res = model.predict( x )
print(np.argmax(res))
print('rec')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from google.colab import files
from io import BytesIO
from PIL import Image

model19 = keras.applications.VGG19(
    include_top=True,
    weights="imagenet",
    input_tensor=None,
    input_shape=None,
    pooling=None,
    classes=1000,
    classifier_activation="softmax",
    name="vgg19",
)

uploaded = files.upload()
img = Image.open(BytesIO(uploaded['ex224.jpg']))
plt.imshow( img )

# приводим к входному формату VGG-сети
img = np.array(img)
# BGR
x = keras.applications.vgg16.preprocess_input(img)
print(x.shape)
# output format: (batch_size, height, width, channels)
x = np.expand_dims(x, axis=0)

# прогоняем через сеть
res = model19.predict( x )
print(np.argmax(res))
print('rec')
```
