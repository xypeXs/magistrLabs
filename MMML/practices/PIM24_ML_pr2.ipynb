```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

np.random.seed(0)
tf.random.set_seed(0)

def generate_sin_data(n_points=1000):
    x = np.linspace(-2*np.pi, 2*np.pi, n_points)
    y = np.sin(x)
    return x.reshape(-1,1), y.reshape(-1,1)

def build_and_train(hidden_units, x_train, y_train, epochs, verbose):
    model = tf.keras.Sequential([
        tf.keras.layers.Dense(hidden_units, activation='tanh', input_shape=(1,)),
        tf.keras.layers.Dense(hidden_units, activation='tanh', input_shape=(hidden_units,)),
        tf.keras.layers.Dense(1, activation='linear')
    ])
    model.compile(optimizer='adam', loss='mse')
    model.fit(x_train, y_train, epochs=epochs, verbose=verbose)
    return model

def plot_approx_and_activations(model, x_plot, y_true, n_neurons_to_plot=10):
    y_pred = model.predict(x_plot)

    plt.figure(figsize=(8,4))
    plt.plot(x_plot.flatten(), y_true.flatten(), label='sin', linewidth=2)
    plt.plot(x_plot.flatten(), y_pred.flatten(), label='approx', linewidth=2)
    plt.title(f'approximaiton (hidden layer neurons = {model.layers[0].units})')
    plt.xlabel('x'); plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    x_input = tf.keras.Input(shape=(1,))
    hidden_output = model.layers[0](x_input)
    activations_model = tf.keras.Model(inputs=x_input, outputs=hidden_output)

    activations = activations_model.predict(x_plot)
    hidden_units = activations.shape[1]
    n_plot = min(n_neurons_to_plot, hidden_units)

    plt.figure(figsize=(10,6))
    for i in range(hidden_units - n_plot, hidden_units):
        plt.plot(x_plot.flatten(), activations[:, i], linestyle='--', label=f'neuron {i}')
    plt.title(f'activation funcs of first {n_plot} neurons (units={hidden_units})')
    plt.xlabel('x'); plt.ylabel('activation func')
    plt.legend(loc='upper right', ncol=2, fontsize='small')
    plt.grid(True)
    plt.tight_layout()
    plt.show()


if __name__ == '__main__':
    x_train, y_train = generate_sin_data(1000)
    x_test, y_test = x_train, y_train

    hidden_sizes = [10, 50, 100, 500]
    for units in hidden_sizes:
        print(f'\n=== train with {units} hidden layer neurons ===')
        model = build_and_train(units, x_train, y_train, epochs=200, verbose=1)
        plot_approx_and_activations(model, x_test, y_test, n_neurons_to_plot=10)
```
